### Tokenize the words and remove stopwords

```python
!pip install bs4
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.gutenberg.org/files/1513/1513-h/1513-h.htm'
response = requests.get(url)
document = response.text
```

```python
#collapse-hide
response.text
```

```python
document = BeautifulSoup(response.text, "html.parser")
```

```python
humtext = document.find('div', attrs={'class':'chapter'}).text
```

```python
for i in ['\n','[',']','’','”','“']:
    humtext = humtext.replace(i,' ')
```

```python
print(humtext)
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
nltk.download('punkt')
```

```python
#collapse-show
sent = sent_tokenize(text)
print(sent)
```

```python
print(sent)
```

```python
print(word_tokenize(sent[1]))
```

```python
words = []
for s in sent:
    for w in word_tokenize(s):
        words.append(w)
```

```python
print(words)
```

```python
words = word_tokenize(text.lower())
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
from string import punctuation
myStopWords = list(punctuation) + stopwords.words('english')
```

```python
wordsNoStopWords = [w for w in words if w not in myStopWords]
```

```python
print(wordsNoStopWords)
```

### Find the top 20 most frequent words in the play

```python
from nltk.probability import FreqDist
```

```python
nltk.download('stopwords')
myStopWords = list(punctuation) + stopwords.words('english')
```

```python
freq = FreqDist(wordsNoStopWords)
```

```python
print(freq.most_common(20))
# The foloowing punctuation and words still appear:
# '’' ; '—' ; 'may'.
```
